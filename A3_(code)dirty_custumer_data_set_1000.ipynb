```python
import re
def data_num(num):
    new_data=re.findall(r'[0-9]+',str(num))
    if(len(new_data)>0):
        return new_data[0]
    else:
        return np.nan
    
```

```python
data["Age"]=data["Age"].apply(lambda x:data_num(x))
```
